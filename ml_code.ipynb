```python
#import modules
import pandas as pd
import numpy as np
```

```python
#read csv file
zoo=pd.read_csv('../input/zoo-animal-classification/zoo.csv')
classs=pd.read_csv('../input/zoo-animal-classification/class.csv')
```

```python
#setup check up
from learntools.core import binder
binder.bind(globals())
from learntools.machine_learning.ex3 import *

print("Setup Complete")
```

```python
zoo.describe()

```

```python
#preprocess the dataset
zoo.dropna(axis=1)
zoo.columns.duplicated()
```

```python
#join zoo.csv and class.csv to get class_type
df=pd.merge(zoo,classs,how='left',left_on='class_type',right_on='Class_Number')
df.head()
```

```python
#plot number of each class_type
import seaborn as sb
sb.catplot('class_type',data=df,kind="count",height=3,aspect=2)
```

```python

#consider features and target value

X=zoo.drop(['animal_name','class_type'],axis=1)
y=zoo['class_type']
X,y
```

```python
#describing the features and target variable
print(X.describe)
print(y.describe)
```

```python
#split the data into trainig set and testing set
from sklearn.model_selection import train_test_split,cross_val_score
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
#create a model and fit the data
from sklearn import svm
#create svm object
c=svm.SVC(kernel='linear',C=0.5)
#fit the data
c.fit(X_train,y_train)
#predict the new data
y_pred=c.predict(X_test)
```

```python
#print the accuracy score and cross validation score
from sklearn import metrics
print("accuracy:",metrics.accuracy_score(y_test,y_pred))
cross_cv=cross_val_score(c,X,y,cv=3)
print("cv:",cross_cv.mean())

```
